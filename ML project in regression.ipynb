```python
# winsorization-Replacing the the values with most permissble values i..ee  lower and upper bounds

def replace_outliers(dataframe,feature):
    Q1=dataframe[feature].quantile(0.25)
    Q3=dataframe[feature].quantile(0.75)

    IQR=Q3-Q1

    upper_bound=Q3+1.5*IQR  
    lower_bound=Q1-1.5*IQR

    # capping-replacing all higher values with upper bound (75% percentile)
    dataframe[feature]=np.where(dataframe[feature]>upper_bound,upper_bound,dataframe[feature])
    # flooring -replace all lower values with lower bounf (25% percentile)
    dataframe[feature]=np.where(dataframe[feature]<lower_bound,lower_bound,dataframe[feature])
```

```python
replace_outliers(df,"transaction_amount")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency, ttest_ind



# Convert categorical columns if needed
categorical_cols = ['item_name','item_type','transaction_type','received_by']
numerical_cols = ['order_id','item_price','quantity','transaction_amount']

# Function to perform chi-square test for categorical variables
def chi_square_test(df, column, target='time_of_sale'):
    table = pd.crosstab(df[column], df[target])
    chi2, p, dof, expected = chi2_contingency(table)
    return p

# Function to perform t-test for numerical variables
def t_test(df, column, target='time_of_sale'):
    group0 = df[df[target] == 0][column]
    group1 = df[df[target] == 1][column]
    t_stat, p = ttest_ind(group0, group1, nan_policy='omit')
    return p

# Run tests and store results
print("Categorical Features - Chi-Square Test")
for col in categorical_cols:
    p = chi_square_test(df, col)
    print(f"{col}: p-value = {p:.4f} => {'Significant' if p < 0.05 else 'Not Significant'}")

print("\nNumerical Features - T-Test")
for col in numerical_cols:
    p = t_test(df, col)
    print(f"{col}: p-value = {p:.4f} => {'Significant' if p < 0.05 else 'Not Significant'}")
```
